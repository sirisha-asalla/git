Under-Sampling

Build sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

```python
# To keep the data balance
legit_sample=legit.sample(n=492)
```

Concatenating two DataFrames.

```python
df=pd.concat([legit_sample,fraud],axis=0)
```

axis=0 we want to add the records as rows.

Splitting the data into Feature and target.

```python
x=df.drop(['Class'],axis=1)
y=df['Class']
```

```python
x
```

```python
y
```

Split the data into training data and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
#Training the logisticregression model with training data
model.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
from sklearn.metrics import accuracy_score,classification_report
#Accuracy on training data
x_train_prediction=model.predict(x_train)
training_acc=accuracy_score(x_train_prediction,y_train)
print('Accuracy Score on training data:',training_acc)
```

```python
#Accuracy on testing data
x_test_prediction=model.predict(x_test)
testing_acc=accuracy_score(x_test_prediction,y_test)
print('Accuracy Score on testing data:',testing_acc)
```

The Accuracy Score is very much Similar to the training data so,there is not overfitting or underfitting in the model.

```python
y_pred=model.predict(x_test)

c=classification_report(y_test,y_pred)
print('classification_report:\n',c)
a=accuracy_score(y_test,y_pred)
print('Accuracy_score:',a)
```
