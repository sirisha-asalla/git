```python
import ast
## create helper function (anything)
def convert(obj):
    ## add the result in a list.
    l=[]
    ## we have run a loop in the list.
    for i in ast.literal_eval(obj):
    ## literal_eval helps to convert str to list
        l.append(i['name'])
    return l
```

```python
df.iloc[0].genres
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head(2)
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
df.head(2)
```

```python
# Here i am just keeping top 3 cast

def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
df['cast']=df['cast'].apply(convert_cast)
```

```python
df.head(2)
```

```python
import ast

def fetch_director(text):
    try:
        crew_list = ast.literal_eval(text)
        return [crew['name'] for crew in crew_list if crew['job'] == 'Director']
    except (SyntaxError, ValueError):
        return []
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df.head(2)
```

```python
## Sam Worthington
## SamWorthington
def remove_space(word):
    l=[]
    for i in word:
        l.append(i.replace(' ',''))
    return l
```

```python
df['cast']=df['cast'].apply(remove_space)
df['crew']=df['crew'].apply(remove_space)
df['genres']=df['genres'].apply(remove_space)
df['keywords']=df['keywords'].apply(remove_space)
```

```python
df.head(2)
```

```python
# Fill any missing values with an empty string and ensure all columns are treated as strings
columns_to_concat = ['overview', 'genres', 'keywords', 'cast', 'crew']
for column in columns_to_concat:
    df[column] = df[column].fillna('').astype(str)
```

```python
# Concatenate the columns
df['tags'] = df['overview'] + ' ' + df['genres'] + ' ' + df['keywords'] + ' ' + df['cast'] + ' ' + df['crew']
```

```python
import nltk
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return ' '.join(l)

```

```python
new_df['tags']=new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similary=cosine_similarity(vector)
```

```python
similary
```

```python
def recommend(movie):
  index=new_df[new_df['title']==movie].index[0]
  distance=sorted(list(enumerate(similary[index])),reverse=True,key=lambda x:x[1])
  for i in distance[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
recommend('Avatar')
```
